```python
# Importing pandas libraty
import pandas as pd
#importing numpy library
import numpy as np
```

```python
#import dataset 
url = "http://bit.ly/autolib_dataset"
df= pd.read_csv(url, encoding='latin1')

df.head(10)
```

```python
#cheking the data information
df.info()
```

# Checking for data validity and missing values

```python
# Checking if there is any missing value in dataframe as a whole
df.isnull()
#there are two null columns: Display Comment and Scheduled at.
```

```python
# checking for Null columns in the dataset
df.isnull().any()
df.isnull().sum()
#the null columns have the following number of missing values:
#Displayed comment = 4889
#Scheduled at = 4953
```

# Removing columns that will not be in use

```python
# removing cars column since it's the same as Bluecar counter.
df1 = df.drop('Cars', 1)
df1.head(5)
```

```python
#Most popular city for picking up a shared electric cars
city= df1.groupby('City').count()['Bluecar counter']
city.sort_values(ascending=False).head(5)
```

```python
# What is the most popular hour for returning cars overall?
hour = df1.groupby('hour').count()['Bluecar counter']
hour.sort_values(ascending=False).head(5)
```

```python
#create a new column for conbined date and time 
df1[['year','month','day','hour','minute']]=df1[['year','month','day','hour','minute']].astype(str)
df1['newdt']=df1['year']+'-'+df1['month']+'-'+df1['day']+' '+df1['hour']+':'+df1['minute']
df1['newdt']=pd.to_datetime(df1['newdt'])
df1.head()

```

# most popular station

```python
#determining the station with the most popular
Station= df1.groupby('ID').count()['Bluecar counter']
Station.sort_values(ascending=False).head(5)
```

```python
#determining thhe most popular post office
postoffice= df1.groupby('Address').count()['Bluecar counter']
postoffice.sort_values(ascending=False).head(5)
```

# the most popular pickup time

```python
# finding the time where most cars were picked
popular_dates = df1.groupby('newdt').count()['Bluecar counter']
popular_dates.sort_values(ascending=True).head(10)
```

# the most popular drop off time

```python
#assuming that cars in parking are the ones dropped off, we can make the excecute the following
# Time with the highest number means many vehicles were dropped of while with the least 
# number of slots means that it is the least drop off time. 
popular_drops = df1.groupby('newdt').count()['Slots']
popular_drops.sort_values(ascending=False).head(10)
```

# most popular time using Utilib and utilib 1.4

```python
#finding most pickup time using utilib
popular_in_utilib = df1.groupby('newdt').count()['Utilib counter']
popular_in_utilib.sort_values(ascending=False).head(10)
```

```python
#finding most pickup time using utilib 1.4
popular_in_utilib4 = df1.groupby('newdt').count()['Utilib 1.4 counter']
popular_in_utilib4.sort_values(ascending=False).head(10)
```
